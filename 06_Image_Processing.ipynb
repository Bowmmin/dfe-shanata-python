```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('moonir.jpeg')

cv2_imshow(img)
```

```python
img.shape # B,G,R -> 3 channels in which we form human-visible image
```

```python
img[0] # first ROW of image
# python-> B,G,R (instead of RGB python follows BGR)
```

```python
img[0][0] # first pixel on first row
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img2)
```

```python
#hue and saturation 
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(img2)
```
